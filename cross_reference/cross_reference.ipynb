```python
banned = []
check_not_success = []
for screen_name in data['screen_name'][:1000]:
    res = requests.get('https://api.shadowban.io/api/v1/twitter/@{0}'.format(screen_name))
    match = re.search(regex, res.text)
    if match.group(1) != 'Check successful':
        check_not_success.append(screen_name)
    if match.group(2) == 'true':
        banned.append(screen_name)
```
